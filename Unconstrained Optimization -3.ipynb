```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

x = np.arange(-5, 5, 0.1)
y = 3 * x - 2

m = np.arange(-5, 5, 0.1)
b = np.arange(-5, 5, 0.1)

m, b = np.meshgrid(m, b)

f = np.zeros(m.shape)

for j in range(1, m.shape[0]):
    for i in range(1, m.shape[0]):
        f[j][i] = f[j][i] + (m[j][i] * x[i] + b[j][i] - y[i])
    f[j] = f[j] / 2

# Plot the surface.
ax.plot_surface(m, b, f, cmap=cm.coolwarm,linewidth=0, antialiased=False)
```
