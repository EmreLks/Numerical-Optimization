```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

r1 = np.arange(10, 90, 1)
r2 = np.arange(10, 90, 1)

r1, r2 = np.meshgrid(r1, r2)

f = np.pi * r1 ** 2 + np.pi * r2 ** 2;

# Plot the surface.
ax.plot_surface(r1, r2, f, cmap=cm.coolwarm,linewidth=0, antialiased=False)

```
